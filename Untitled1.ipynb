```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('trade_new.csv')
df.head()
```

```python
fig, ax = plt.subplots(10*5,1)
for i in np.sort(df['SITC 1 DIGIT'].unique()):
    data = df[df['SITC 1 DIGIT'] == i].groupby('YEAR')[['IMPORT (MILLION RM)', 'EXPORT (MILLION RM)']].sum().reset_index() 
    sns.heatmap(data.corr(), annot = True, ax = ax[i])
```

```python
data = df[df['SITC 1 DIGIT'] == 1].groupby('YEAR')[['IMPORT (MILLION RM)', 'EXPORT (MILLION RM)']].sum().reset_index()
data1 = df[df['SITC 1 DIGIT'] == 2].groupby('YEAR')[['IMPORT (MILLION RM)', 'EXPORT (MILLION RM)']].sum().reset_index()
fig, ax = plt.subplots(2,1) 
sns.heatmap(data.corr(), annot = True, ax = ax[0])
sns.heatmap(data1.corr(), annot = True, ax= ax[1])
```
